```python
# 5. Crea un dataframe con TODOS os episodios (de todas as páxinas)
```

```python
import pandas as pd
df_list = []
numPaginas = response.json()['info']['pages']
for numPagina in range(1,numPaginas+1):
    payload = {'page': numPagina}
    response = requests.get(endpoint_episodios,params = payload)
    data = response.json()
    df = pd.DataFrame(data['results'])
    df_list.append(df)

result = pd.concat(df_list)
dfUnico = pd.DataFrame(result)
print(dfUnico)
dfUnico.describe()

```
